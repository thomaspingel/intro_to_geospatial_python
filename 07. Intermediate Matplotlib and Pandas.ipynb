```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

One of the most handy features of Matplotlib is the ability to subplot, or create multiple plots within the same frame.

The plt.subplot() function will prepare the space.  You provide the rows, columns, and the number of the plot.

For example:
plt.subplot(3,2,1) would create a space for six plots, and activate the first space.:

1      2

3      4

5      6

```python
# From https://matplotlib.org/examples/pylab_examples/subplot_demo.html

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'ko-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

```python
# From https://matplotlib.org/examples/pylab_examples/subplot_demo.html

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 2, 1)
plt.plot(x1, y1, 'ko-')

plt.subplot(2, 2, 2)
plt.plot(x2, y2, 'r.-')

plt.subplot(2, 2, 3)
plt.plot(x2, y2, 'g-.')

plt.subplot(2, 2, 4)
plt.plot(x2, y2, 'yo--')

plt.show()
```

```python
# You can also create subplots in multiple "spaces" that overlap.  For instance:

x = np.linspace(0,10)
y = x**2

# The first specification is making a space 2 rows by 1 column.  So far so good.
plt.subplot(2,1,1)
plt.plot(x,y)

# But these next two are creating, in the same figure, a space 2x2, and filling positions 3 and 4 
# (bottom left and right, respectively).  The result is a sort of "hybrid" space that can be very 
# useful!
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.show()
```

## Deleting rows and columns

```python
# Although not often done, columns can be deleted with the DEL operator, which can also delete variables in memory:

x = pd.DataFrame(np.random.rand(5,3),columns=['a','b','c'])
del x['b']
x.head()
```

```python
# An example of using the DEL operator to delete an object in memory
x = 5
del x
print(x)
```

```python
# The preferred method of deleting uses the DROP method:

# axis=1 tells Pandas that you want to drop a column (Rows and columns COULD have the same ID!)

x = pd.DataFrame(np.random.rand(5,3),columns=['a','b','c'])
x = x.drop('b',axis=1)
x.head()
```

```python
# Methods like DROP have an optional parameter to avoid the "df = df.drop(..") style syntax: The INPLACE parameter:
x = pd.DataFrame(np.random.rand(5,3),columns=['a','b','c'])
x.drop('b',axis=1,inplace=True)
x.head()
```

```python
## Selection with loc vs iloc.
# LOC will grab based on the index VALUE: Which row is MARKED "3"
# ILOC will grab based on POSITION and ignore value: Which row is in position 3?

x = pd.DataFrame(np.random.rand(5,3),columns=['a','b','c'])
x.drop(2,axis=0,inplace=True)
print(x.head())

print('\n')

print('This is iloc:3')
print(x.iloc[3,:])

print('\n')

print('This is loc:3')
print(x.loc[3,:])
```

## Reading, Writing, and Repair

```python
# The datafile we have for Fisher's Iris Data is very dense, and should have metadata.
# Lets reformat it using the skills we've developed as a simple "repair" example.
df = pd.read_csv('Fisher.csv')
df.head()
```

```python
# We can rename the columns to something a little more familiar:
df.rename(columns={'PW':'petal_width'},inplace=True)
df.head()
```

```python
# Play along!  Rename PL, SW, SL to petal_width, sepal_width, and sepal_length.
```

```python
# We can use a selector to change the values for 0,1,2 to their species name:

df.loc[df['Type']==0,'Type'] = 'setosa'
df.head()
```

```python
# Play along!  Change Type of 1 to 'virginica' and 2 to 'versicolor'.  Rename 'Type' to 'species'.
```

```python
# Here's the whole repair:

df = pd.read_csv('Fisher.csv')
typefix = {0:'setosa',1:'virginia',2:'versicolor'}
colfix = {'Type':'species','PW':'petal_width','PL':'petal_length','SW':'sepal_width','SL':'sepal_length'}
for item in typefix:
    df.loc[df['Type']==item,'Type'] = typefix[item]
df.rename(columns=colfix,inplace=True)
df.head()
```

```python
# Now that the repair is done, let's write it out!

# Pandas uses a "df.to_*" syntax to write out the data:

df.to_csv('Fisher_nice.csv')
```

```python
# If you don't want the index value to be written, use this to suppress it:

df.to_csv('Fisher_nice.csv',index=False)
```
