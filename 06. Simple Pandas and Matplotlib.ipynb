## Line Plots of NumPy Arrays (mathematical functions)

```python
# You can generate some simple figures by using numpy to create a range of numbers, and then do some caluclation.

t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2*np.pi*t)
plt.plot(t, s)

# Notice that the x and y axis are not equally sized! - Matplotlib is scaling these so that all the data 
# fits on the screen.  You sometimes have to be careful in your interpretation as a result!
```

```python
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2*np.pi*t)
plt.plot(t, s)

# You can annotate your graphs this way
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.grid(True)

# And save them with something like this
plt.savefig("test.png")
```

```python
# You can use a similar array to select your data:

# In which rows does is the System Mode set to compressorCoolState1On 
# (i.e., the air conditioning is on)

ac_rows = df['System Mode'] =='compressorCoolStage1On'

ac_rows
```

```python
# Remember that True is like 1, and False is like 0.  We can find out what percentage of time
# the air conditioner was running this way:

ac_rows = df['System Mode'] =='compressorCoolStage1On'
prc_on = 100 * sum(ac_rows) / len(ac_rows)
print('The air conditioner was running',prc_on,'percent of the time.')

# The SUM of all those Trues is how many times the AC was on.
# The LENGTH of the record is how many total rows I have.
# So the SUM divided by the LENGTH is the proportion of time the AC was on.
# This multiplied by 100 is the percentage.
```

## Histograms

```python
# Another handy plot is the histogram:

# Create 1000 normally distributed random numbers
x = np.random.normal(loc=0,scale=1,size=1000)

# And plot them:  
# (Note, I'm collecting the output from this function to a THROWAWAY VARIABLE)
_ = plt.hist(x,bins=100)
```

```python
# And plot them:  
# (Note, I'm collecting the output from this function to a THROWAWAY VARIABLE)
_ = plt.hist(x,bins=20)
```

```python
# I'd like to do the same with my real data, but there's a problem:
x = df['Current Temp (F)']
_ = plt.hist(x)
```

```python
# The problem is, I'm missing some data, and histogram isn't sure what to do.

# See them there at the bottom?  These are called "nans" or "Not a Number" and they correspond to missing values.

x
```

### Removing NaNs in your data

```python
# We can find them, and "remove" them by keeping only rows that have valid values:

x = df['Current Temp (F)']
x = x[~np.isnan(x)]
_ = plt.hist(x,bins=50)
```

### Specifying bin boundaries instead of number of bins

```python
x = df['Current Temp (F)']
x = x[~np.isnan(x)]

# You can also pass hist a "bins" argument to specify the boundaries
_ = plt.hist(x,bins=np.arange(68.5,80.5,1))
```
