```python
# See http://www.math.uah.edu/stat/data/Fisher.html
df = pd.read_csv('http://www.math.uah.edu/stat/data/Fisher.csv')
```

```python
df.head()
```

```python
np.unique(df.Type)
```

```python
plt.plot(df.PW[df.Type==0],df.PL[df.Type==0],'r.',label='Setosa')
plt.plot(df.PW[df.Type==1],df.PL[df.Type==1],'g.',label='Verginica')
plt.plot(df.PW[df.Type==2],df.PL[df.Type==2],'b.',label='Versicolor')
plt.legend(loc='upper left')
```

```python
plt.hist(df.PL)
```

```python
import numpy as np
import seaborn as sns

sns.set_style('whitegrid')
sns.kdeplot(df.PL, bw=0.5)
```

```python
import seaborn as sns
sns.set(style="ticks")

df = pd.read_csv('http://www.math.uah.edu/stat/data/Fisher.csv')

# Show the results of a linear regression within each dataset
sns.lmplot(x="PW", y="PL", data=df)
```
